## Data Description

**Police recorded crime:**

The dataset is structured in rows and columns. Each row contains a single count of crime, and each column provides information about that count, including the time period, offense type, and the area it applies to. 

Below are the descriptions for each column:

- **Financial Year**: Combined with the **Financial Quarter** column, this identifies the period during which offenses were initially recorded, which is when the crime was reported to the police. Each financial year runs from April to March.
  
- **Financial Quarter**: Combined with the **Financial Year** column, this identifies the period during which offenses were recorded by the police. Quarter 1 runs from April to June, Quarter 2 from July to September, Quarter 3 from October to December, and Quarter 4 from January to March.
  
- **Force Name**: This column identifies the police force area within which the offense was recorded, generally corresponding to the area where the offense occurred.
  
- **Offense Description**: This column describes the offense associated with each **Offence Code**.

- **Offence Group**: This column identifies the group of offenses within which the **Offence Code** falls. Each group consists of multiple **Offence Subgroups**, which in turn contain specific **Offence Codes**.

- **Offence Subgroup**: This column specifies the sub-group within which the **Offence Code** falls, containing a set of related offenses.

- **Offence Code**: This column provides the specific code used by the police and the Home Office to classify offenses.

- **Number of Offences**: This column contains the total number of police-recorded crimes for the specified **Offence Code**, **CSP Name/Force Name**, and time period (Financial Year andFinancial Quarter).

## Step 1: Read and Merge Excel sheets (Data Creation)

```python
# Read the Excel file and get all sheet names
file = "Police_Recorded_Data.xlsx"
data = pd.ExcelFile(file)

# Ignore the first two sheets and merge the remaining sheets
df = pd.concat([pd.read_excel(data, sheet_name=sheet) for sheet in data.sheet_names[2:]])

# Display the first 5 rows of the combined DataFrame
df.head()
```

## Step 3: Convert Financial Year and Quarter to Dates

```python
# Convert Financial Year and Quarter to Dates
def convert_financial_year_quarter(row):
    financial_year = row['Financial Year']
    quarter = row['Financial Quarter']
    start_year = int(financial_year.split('/')[0])
    month_mapping = {1: '04', 2: '07', 3: '10', 4: '01'}
    month = month_mapping[quarter]
    year = start_year if quarter != 4 else start_year + 1
    return f'{year}-{month}-01'
```

```python
df['Date'] = pd.to_datetime(df.apply(convert_financial_year_quarter, axis=1))
df[['Date']]
```

## Trend Analysis:
__Research Question 1: How have overall crime trends changed over time?:__

```python
# Analyzing the overall trend in all types of offences
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Number of Offences', data=df.groupby('Date')['Number of Offences'].sum().reset_index(), linewidth=2, color='b')
plt.title('Overall Crime Trends', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Offences', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Observation
This analysis shows fluctuations in overall crime trends over time, with notable increases in recent years. Suggesting a need for continuous adaptation of law enforcement strategies to address changing crime patterns.

__Research Question 2: What are the top 5 most common offences?__

```python
# Investigate the Top 5 Most common Crimes
top_offences = df.groupby('Offence Description')['Number of Offences'].sum().nlargest(5)
plt.figure(figsize=(14, 7))
sns.barplot(x=top_offences.index, y=top_offences.values, palette='viridis', hue=top_offences.index)
plt.title('Top 5 Most Common Offences', fontsize=16)
plt.xlabel('Offence Description', fontsize=12)
plt.ylabel('Number of Offences', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Observation:
The top five most common offences include: Assault without Injury, Assault with Injury, Other Theft, Shoplifting, and Fraud Offences from records.
Identifying the most common offences helps in prioritizing resources and developing targeted prevention strategies. The bar chart provides a clear visualization of which crimes are most prevalent.

__Research Question 3: How have fraud offences evolved overtime?__

```python
# Analyzing Fraud Offences Overtime
fraud_df = df[df['Offence Group'] == 'Fraud offences']
fraud_trend = fraud_df.groupby('Date')['Number of Offences'].sum()
plt.figure(figsize=(12, 6))
fraud_trend.plot()
#plt.axvline(x=pd.to_datetime('2014-04-01'), color='red', linestyle='--', label='Introduction of Action Fraud')
plt.title('Trend in Fraud Offences')
plt.xlabel('Date')
plt.ylabel('Number of Fraud Offences')
plt.legend()
plt.show()
```

__Research Question 4: Is there a correlation between different types of crimes?__

```python
# Research Question 4: Is there a correlation between different types of crimes?
crime_pivot = df.pivot_table(index='Date', columns='Offence Group', values='Number of Offences', aggfunc='sum')
correlation_matrix = crime_pivot.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Between Crime Types', fontsize=16)
plt.tight_layout()
plt.show()
```

#### Observations:

The correlation matrix heatmap reveals potential relationships between different types of crimes. From the HeatMap, there is a strong positive correlation between certain crime types, such as property(Robbery, Theft, Criminal Damage and Arson) and violent crimes(Violent against the person, Sexual Offence, Possession of weapons), suggesting that these crimes often occur together or influence each other.

These correlations can help identify underlying factors, such as economic hardship or social conditions, that contribute to specific crime patterns.
There are correlations between different types of crimes, suggesting potential underlying factors.

This information can be valuable for understanding how different criminal
activities might be interconnected and for developing comprehensive crime prevention strategies.

__Research Question 6: How do crime rates vary across different police forces?__

```python
# Research Question 6: How do crime rates vary across different police forces?
force_crime_rates = df.groupby('Force Name')['Number of Offences'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(15, 8))
sns.barplot(x=force_crime_rates.index, y=force_crime_rates.values, palette='colorblind', hue=force_crime_rates.index)
plt.title('Crime Rates Across Police Forces', fontsize=16)
plt.xlabel('Force Name', fontsize=15)
plt.ylabel('Total Number of Offences', fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Observations:
Crime rates vary significantly across different police forces, with Metropolitan Police topping, followed by Action Fraud, Greater Manchester, west midlands and West Yorkshire


__Research Question 7: What is the trend in drug-related offences?__

__Research Question 9: What is the break down of crimes by offence group?:__

```python
# Group by offence category and year
offense_category_trend = df.groupby(['Offence Group', df['Date'].dt.year])['Number of Offences'].sum().unstack()

# Plot trends by offence category
plt.figure(figsize=(16, 10))
offense_category_trend.T.plot(ax=plt.gca(), linewidth=2, marker='o')

# Customize the plot
plt.title('Trends in Crimes by Offence Group', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Offence Group', bbox_to_anchor=(1.05, 1), loc='upper left')

# Improve layout and display
plt.tight_layout()
plt.show()
```

The Distribution of different offence groups based on  number of crimes over the years
